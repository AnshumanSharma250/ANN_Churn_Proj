```python
!pip install pandas
!pip install scikit-learn
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler , LabelEncoder
import pickle
```

```python
## load the dataset
df = pd.read_csv(r'C:\Projects\ANN-DL_Bank_Proj\Churn_Modelling.csv')
df.head()
```

```python
df=df.drop(['RowNumber','CustomerId','Surname'],axis=1)
df
```

```python
## geographical and gender ar ecategorical 
label_encoder_gender = LabelEncoder()
df['Gender'] = label_encoder_gender.fit_transform(df['Gender'])
df
```

```python
df.info()
```

```python
## here now we are applying OHE on geogrpahic columns
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
geo_encoder = ohe.fit_transform(df[['Geography']])
geo_encoder
```

```python
ohe.get_feature_names_out(['Geography'])
```

```python
geo_encoded_df = pd.DataFrame(geo_encoder.toarray(),columns =ohe.get_feature_names_out(['Geography']) )
```

```python
## combine the ohe columsn in original data
df = pd.concat([df.drop('Geography',axis=1),geo_encoded_df],axis=1)
```

```python
df.head()
```

```python
## save the encoder and scaler
with open('label_encoder_gender.pkl','wb') as file:
    pickle.dump(label_encoder_gender,file)

with open('ohe','wb') as file:
    pickle.dump(ohe,file)
```

```python
## divide the data in dep and indep features
x = df.drop('Exited',axis=1)
y=df['Exited']

## splitting in train g ansd testing daata
x_train ,x_test ,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)

## scale these features
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
with open('scaler.pkl','wb') as file:
    pickle.dump(scaler,file)
```

# ANN Implementation

```python
!pip install tensorflow
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping , TensorBoard
import datetime
```

```python
## Build ANN model
model = Sequential([
    Dense(64,activation='relu',input_shape = (x_train.shape[1],)), ## HL1 connected to input layer
    Dense(32,activation='relu'),##HL2
    Dense(1,activation='sigmoid') ## o/p layer
      
])
```

```python
model.summary()
```

```python
import tensorflow
opt = tensorflow.keras.optimizers.Adam(learning_rate=0.01)

## note that we can s etup oour own loss function and then feed it down for compiling
```

```python
## compile the model
model.compile(optimizer=opt,loss = "binary_crossentropy",metrics = ['accuracy'])
```

```python
## setup the tensorboard
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard

log_dir = "logs/fit" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorflow_callback = TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
## tensorboard is a visualization tool provided by tensorflow
##  Visualizes model graph/architecture
```

```python
## set up early stopping
early_stopping_callback = EarlyStopping(monitor='val_loss',patience=10,restore_best_weights=True)
```

```python
## train the model
history = model.fit(
    x_train,y_train,validation_data=(x_test,y_test),epochs=100,
    callbacks = [tensorflow_callback,early_stopping_callback]
)
```

```python
model.save('model.h5')
```
